```python
# Import required libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset from a CSV file
df = pd.read_csv('/content/drive/MyDrive/Mini/bank_loan_prediction.csv')
```

```python
# Split the dataset into features and target variable
X = df.drop('Loan Status', axis=1)
y = df['Loan Status']
#y = y.apply(pd.to_numeric)
```

```python
# Encode the target variable
le = LabelEncoder()
y = le.fit_transform(y)


# encode the categorical column
#df['Batch Enrolled'] = le.fit_transform(df['Batch Enrolled'])

df['Grade'] = le.fit_transform(df['Grade'])
df['Sub Grade'] = le.fit_transform(df['Sub Grade'])
df['Employment Duration'] = le.fit_transform(df['Employment Duration'])
df['Verification Status'] = le.fit_transform(df['Verification Status'])
#df['Payment Plan'] = le.fit_transform(df['Payment Plan'])
df['Loan Title'] = le.fit_transform(df['Loan Title'])
df['Application Type'] = le.fit_transform(df['Application Type'])
df['Initial List Status'] = le.fit_transform(df['Initial List Status'])
X = df
# df[''] = le.fit_transform(df[''])
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
# Create a random forest classifier with 100 trees
clf = RandomForestClassifier(n_estimators=50, random_state=25)
```

```python
# Train the classifier on the training data
clf.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = clf.predict(X_test)
```

```python
# Evaluate the performance of the classifier
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Make predictions on the test data
y_pred = clf.predict(X_test)
```

```python
# Calculate precision, recall, and F1-score
precision = precision_score(y_test, y_pred, pos_label=1)
recall = recall_score(y_test, y_pred, pos_label=1)
f1 = f1_score(y_test, y_pred, pos_label=1)

print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-score: {f1:.2f}')
```

```python
# Get feature importances
feature_importances = clf.feature_importances_

# Identify sensitive features (e.g., with importance above a threshold)
sensitive_features = X.columns[feature_importances > 0.1]  # Adjust the threshold as needed

# Extract sensitive features from the original dataset
sensitive_df = df[sensitive_features]

# Save the sensitive dataset to a new CSV file
sensitive_df.to_csv('sensitive_dataset.csv', index=False)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import csv
import rsa

# Load the dataset from a CSV file
df = pd.read_csv('/content/drive/MyDrive/Mini/bank_loan_prediction.csv')

# Split the dataset into features and target variable
X = df.drop('Loan Status', axis=1)
y = df['Loan Status']

# Encode the target variable
le = LabelEncoder()
y = le.fit_transform(y)

# Encode the categorical columns
cat_cols = ['Grade', 'Sub Grade', 'Employment Duration', 'Verification Status', 'Loan Title', 'Application Type', 'Initial List Status']
for col in cat_cols:
    X[col] = le.fit_transform(X[col])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a random forest classifier with 10 trees
clf = RandomForestClassifier(n_estimators=10, random_state=42)

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = clf.predict(X_test)

# Evaluate the performance of the classifier
print("Accuracy:", accuracy_score(y_test, y_pred))

# Generate RSA key pair
(public_key, private_key) = rsa.newkeys(2048)

# Print public and private keys
print("Public key: ", public_key)
print("Private key: ", private_key)

# Read data from CSV file
with open('/content/sensitive_dataset.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    data = [row for row in csv_reader]

# Find patterns and modify sensitive data
modified_data = []
for row in data:
    modified_row = []
    for col in row:
        if col.isnumeric():
            # Encrypt numeric values using RSA
            encrypted_value = rsa.encrypt(str(col).encode(), public_key)
            modified_row.append(encrypted_value)
        else:
            # Leave non-numeric values unchanged
            modified_row.append(col)
    modified_data.append(modified_row)

# Write modified data to CSV file
with open('output.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    for row in modified_data:
        csv_writer.writerow(row)

# Read modified data from CSV file and decrypt numeric values
with open('output.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    modified_data = [row for row in csv_reader]

for row in modified_data:
    for col in row:
        if isinstance(col, bytes):
            # Decrypt encrypted values using RSA
            decrypted_value = rsa.decrypt(col, private_key).decode()
            print(f'Encrypted value: {col}')
            print(f'Decrypted value: {decrypted_value}')
        else:
            print(f' {col}')
```
