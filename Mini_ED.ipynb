```python
import csv
import rsa

# Generate RSA key pair
(public_key, private_key) = rsa.newkeys(2048)


# Print public and private keys
print("Public key: ", public_key)
print("Private key: ", private_key)

# Load data from CSV file
with open('/content/sensitive_dataset.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    data = [row for row in csv_reader]
print(data)

# Find patterns and modify sensitive data
modified_data = []
for row in data:
    modified_row = []
    for col in row:
        if col.isnumeric():
            # Encrypt numeric values using RSA
            encrypted_value = rsa.encrypt(str(col).encode(), public_key)
            modified_row.append(encrypted_value)
        else:
            # Leave non-numeric values unchanged
            modified_row.append(col)
    modified_data.append(modified_row)

# Write modified data to CSV file
with open('output.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    for row in modified_data:
        csv_writer.writerow(row)
```

```python
import csv
import os
from Crypto.Cipher import Threefish

# Generate a random encryption key
key = os.urandom(64)

# Open CSV file for reading and writing
with open('/content/drive/MyDrive/Mini/diabetes.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    
    # Open new CSV file for writing
    with open('encrypted_data.csv', 'w', newline='') as encrypted_file:
        csv_writer = csv.writer(encrypted_file)
        
        # Create a Threefish cipher object
        cipher = Threefish.new(key, Threefish.MODE_CTR)
        
        # Iterate through each row in the input CSV file
        for row in csv_reader:
            modified_row = []
            # Iterate through each column in the current row
            for col in row:
                if col.isnumeric():
                    # Encrypt numeric values using Threefish
                    encrypted_value = cipher.encrypt(str(col).encode())
                    modified_row.append(encrypted_value)
                else:
                    modified_row.append(col)
            
            # Write modified row to the output CSV file
            csv_writer.writerow(modified_row)

# Open encrypted CSV file for reading
with open('encrypted_data.csv', 'r') as encrypted_file:
    csv_reader = csv.reader(encrypted_file)
    
    print("Encrypted File")
    # Iterate through each row in the encrypted CSV file
    for row in csv_reader:
        # Iterate through each column in the current row
        for col in row:
            # Print encrypted value
            print(col)

# Open encrypted CSV file for reading and writing
with open('encrypted_data.csv', 'r') as encrypted_file, open('decrypted_data.csv', 'w', newline='') as decrypted_file:
    csv_reader = csv.reader(encrypted_file)
    csv_writer = csv.writer(decrypted_file)
    
    # Create a new Threefish cipher object
    cipher = Threefish.new(key, Threefish.MODE_CTR)
    
    # Iterate through each row in the encrypted CSV file
    for row in csv_reader:
        modified_row = []
        # Iterate through each column in the current row
        for col in row:
            if col.isnumeric():
                # Decrypt numeric values using Threefish
                decrypted_value = cipher.decrypt(col)
                modified_row.append(decrypted_value.decode())
            else:
                modified_row.append(col)
        
        # Write modified row to the output CSV file
        csv_writer.writerow(modified_row)

# Open decrypted CSV file for reading
with open('decrypted_data.csv', 'r') as decrypted_file:
    csv_reader = csv.reader(decrypted_file)
    
    print("\nDecrypted File")
    # Iterate through each row in the decrypted CSV file
    for row in csv_reader:
        print(row)
```
